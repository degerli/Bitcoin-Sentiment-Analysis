## More preprocessing

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import re
```

```python
nltk.download('punkt')
```

```python
emoji_pattern = re.compile("["
         u"\U0001F600-\U0001F64F"  
         u"\U0001F300-\U0001F5FF" 
         u"\U0001F680-\U0001F6FF"  
         u"\U0001F1E0-\U0001F1FF"  
         u"\U00002702-\U000027B0"
         u"\U000024C2-\U0001F251"
         "]+", flags=re.UNICODE)
```

```python
import string
import re
def clean_tweetdata(tweet):
 
    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(tweet)
    tweet = re.sub(r':', '', tweet)
    tweet = re.sub(r'‚Äö√Ñ¬∂', '', tweet)
    tweet = re.sub(r'[^\x00-\x7F]+',' ', tweet)
    tweet = emoji_pattern.sub(r'', tweet)
    filtered_tweet = [w for w in word_tokens if not w in stop_words]
    filtered_tweet = []
    for w in word_tokens:
        if w not in stop_words and w not in string.punctuation:
            filtered_tweet.append(w)
    return ' '.join(filtered_tweet)
```

```python
for ind,i in enumerate(cleantweets):
    cleantweets[ind]=clean_tweetdata(i)
```

## Now we willl use TextBlob library to perform sentiment analysis

```python
from textblob import TextBlob
```

```python
sentiments=[]
subjectivity=[]
for i in tweetdatadatewise['Tweet']:
    textblb = TextBlob(i)
    sentiments.append(textblb.polarity)
    subjectivity.append(textblb.subjectivity)
```

```python
pospol=0
negpol=0
nopol=0
subcount=0
for i in sentiments:
        if (i ==0):
            nopol+=1
        elif(i>0):
            pospol+=1
        elif(i<0):
            negpol+=1
for i in subjectivity:
    if(i>0.5):
        subcount+=1
print(' Zero polarities are '+ str(nopol))
print(' Positive polarities are '+ str(pospol))
print(' Negative polarities are '+str(negpol))
print('Number of subjective tweets are : '+ str(subcount))
```

### Number of tweets with subjectivity>0.5 are 192

```python
print("Average sentiment is: "+str(sum(sentiments)/len(sentiments)))
```

```python
import matplotlib.pyplot as plt
pols=['Neutral', 'Positive', 'Negative']
polcount=[nopol,pospol,negpol]
plt.pie(polcount, labels = pols,autopct='%1.2f%%')
```

## We will consider the tweets with subjectivity>0.5 because less subjective tweets tend to be facts and news rather than opinions.

```python
sentimentday=[]
tweetdata_filtered = tweetdatadatewise[tweetdatadatewise['Subjectivity'] >= 0.5]
sentiment_filtered=list(tweetdata_filtered.Sentiment)
times = pd.DatetimeIndex(tweetdata_filtered.Date)
grouped = tweetdata_filtered.groupby(times.date).mean()
#for i in range(8):
 #   sentimentday.append(sum(tweetdatadatewise['Sentiment'][i*100:(i+1)*100])/100)
```

## Now, forecasting using ARIMA

```python
from statsmodels.tsa.arima_model import ARIMA
from sklearn.metrics import mean_squared_error
```

```python
history = [x for x in sentimentday]

model = ARIMA(history, order=(1,1,0))
model_fit = model.fit(disp=0)
output = model_fit.forecast()
yhat = output[0]
print( "Next day sentiment is forecasted to be: "+str(yhat))
```

```python
plt.plot(range(0,8),sentimentday)
plt.scatter(8,yhat,color='red')
```

## Calculating Mean Squared Error of ARIMA

```python
train, test = sentiment_filtered[0:180], sentiment_filtered[180:len(sentiment_filtered)]
history = [x for x in train]
predictions = list()
for t in range(len(test)):
	model = ARIMA(history, order=(5,1,0))
	model_fit = model.fit(disp=0)
	output = model_fit.forecast()
	yhat = output[0]
	predictions.append(yhat)
	obs = test[t]
	history.append(obs)
	print('predicted=%f, expected=%f' % (yhat, obs))
```

```python
error = mean_squared_error(test, predictions)
print('Test MSE: %.3f' % error)
plt.plot(test)
plt.plot(predictions, color='red')
plt.show()
```

## Test MSE is calculated to be 0.145

### Finding co-variance matrix

```python
from numpy import cov
```

```python
covar1=cov(bitcoin['24h High (USD)'],sentimentday)
covar2=cov(bitcoin['24h Low (USD)'],sentimentday)
covar3=cov(bitcoin['24h Open (USD)'],sentimentday)
covar4=cov(bitcoin['Closing Price (USD)'],sentimentday)
```

```python
covar1,covar2,covar3,covar4
```

### In the case where covariance matrix is positive, we can expect that the variables change in the same direction while if negative, the change will be in opposite direction

## Karl Pearson's coefficient for correlation

```python
from scipy.stats import pearsonr
```

```python
correlation1, x=pearsonr(bitcoin['24h High (USD)'],sentimentday)
correlation2, x=pearsonr(bitcoin['24h Low (USD)'],sentimentday)
correlation3, x=pearsonr(bitcoin['24h Open (USD)'],sentimentday)
correlation4, x=pearsonr(bitcoin['Closing Price (USD)'],sentimentday)
```

```python
correlation1,correlation2,correlation3,correlation4
```

# Now we calculate Spearman's correlation coefficient

```python
from scipy.stats import spearmanr
```

```python
correlation1, x=spearmanr(bitcoin['24h High (USD)'],sentimentday)
correlation2, x=spearmanr(bitcoin['24h Low (USD)'],sentimentday)
correlation3, x=spearmanr(bitcoin['24h Open (USD)'],sentimentday)
correlation4, x=spearmanr(bitcoin['Closing Price (USD)'],sentimentday)
```

```python
correlation1,correlation2,correlation3,correlation4
```

## Therefore we can say that the sentiment data and the bitcoin data are not much related to each other.
